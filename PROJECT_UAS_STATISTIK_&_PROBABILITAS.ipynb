```python
# TUGAS MATA KULIAH STATISTIK & PROBABILITAS
# Analisis Regresi Linier Berganda - Kasus Biaya Asuransi Kesehatan

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error

print("Koneksi sukses! Library siap digunakan.")

df = pd.read_csv('insurance.csv')

print("\n=== 5 Data Pertama di Dataset ===")
print(df.head())

df['is_male'] = df['sex'].map({'male': 1, 'female': 0})
df['is_smoker'] = df['smoker'].map({'yes': 1, 'no': 0})

df_final = pd.get_dummies(df, columns=['region'], drop_first=True, dtype=int)

print("\n=== Data Setelah Selesai Diolah (Siap Masuk Rumus) ===")
print(df_final.head())


Fitur_X = df_final[['age', 'bmi', 'children', 'is_male', 'is_smoker',
                     'region_northwest', 'region_southeast', 'region_southwest']]
Target_y = df_final['charges']

X_train, X_test, y_train, y_test = train_test_split(Fitur_X, Target_y, test_size=0.2, random_state=42)

print(f"\nJumlah Data Training: {X_train.shape[0]} baris")
print(f"Jumlah Data Testing: {X_test.shape[0]} baris")

model_regresi = LinearRegression()
model_regresi.fit(X_train, y_train)

y_prediksi = model_regresi.predict(X_test)

nilai_r2 = r2_score(y_test, y_prediksi)
nilai_mae = mean_absolute_error(y_test, y_prediksi)
nilai_rmse = np.sqrt(mean_squared_error(y_test, y_prediksi))

print("\n=============================================")
print("          HASIL EVALUASI MODEL REGRESI       ")
print("=============================================")
print(f"R-Squared (R²)         : {nilai_r2:.4f} (atau {nilai_r2*100:.2f}%)")
print(f"Mean Absolute Error (MAE): ${nilai_mae:.2f}")
print(f"Root Mean Squared Error: ${nilai_rmse:.2f}")
print("=============================================")

print("\n=== NILAI PARAMETER RUMUS REGRESI ===")
print(f"Konstanta/Titik Potong (Intercept - Beta 0): {model_regresi.intercept_:.2f}")
print("\nKoefisien Pengaruh (Slopes / Coefficients):")
for kolom, koefisien in zip(Fitur_X.columns, model_regresi.coef_):
    print(f" - {kolom:<18}: {koefisien:.2f}")

# Grafik 1: Perbandingan Tebakan vs Aktual
plt.scatter(y_test, y_prediksi, alpha=0.5, color='teal')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Biaya Asli ($)')
plt.ylabel('Biaya Prediksi Model ($)')
plt.title('Grafik Evaluasi: Biaya Asli vs Prediksi Model')
plt.tight_layout()
plt.savefig('actual_vs_predicted.png')
plt.close()

# Grafik 2: Pengaruh Fitur
coef_df = pd.DataFrame({'Fitur': Fitur_X.columns, 'Koefisien': model_regresi.coef_}).sort_values(by='Koefisien')
plt.barh(coef_df['Fitur'], coef_df['Koefisien'], color='royalblue')
plt.xlabel('Nilai Koefisien Pengaruh')
plt.title('Seberapa Besar Pengaruh Tiap Faktor Terhadap Biaya?')
plt.tight_layout()
plt.savefig('feature_coefficients.png')
plt.close()

print("\n[PROSES SELESAI] Grafik hasil evaluasi telah disimpan secara otomatis!")
```
